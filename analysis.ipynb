```python
from pathlib import Path
import pandas as pd

repo = Path("turkish-hate-speech-dataset-2")
print("Repo exists?", repo.exists(), repo.resolve())

# Aday dosyalar: csv/tsv/xlsx/json
cands = []
for ext in ["*.csv", "*.tsv", "*.xlsx", "*.json", "*.txt"]:
    cands += list(repo.rglob(ext))

print("Candidate files:", len(cands))
print("\n".join(str(p.relative_to(repo)) for p in cands[:30]))

# İçeriğe bakarak "tweet/text" kolonu olanları bul (hızlı tarama)
hits = []
for p in cands:
    try:
        if p.suffix.lower() == ".csv":
            df = pd.read_csv(p)
        elif p.suffix.lower() == ".tsv":
            df = pd.read_csv(p, sep="\t")
        elif p.suffix.lower() == ".xlsx":
            df = pd.read_excel(p)
        else:
            continue
        cols = [str(c).lower() for c in df.columns]
        if any(k in cols for k in ["tweet","text","content"]) or any("tweet" in c or "text" in c for c in cols):
            hits.append((p, df.shape, df.columns.tolist()))
    except Exception:
        pass

print("\nFILES WITH TEXT/TWEET-LIKE COLUMNS:", len(hits))
for p, shape, cols in hits[:10]:
    print("-", p.relative_to(repo), "| shape:", shape, "| cols:", cols)
```

```python
from datasets import load_dataset
import pandas as pd

ds_tw = load_dataset("stefan-it/offenseval2020_tr")
print(ds_tw)
print(ds_tw["train"][0])
```

```python
def parse_fasttext(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    raw = df["text"].astype(str)

    # __label__NOT / __label__OFF
    df["Label"] = raw.str.extract(r"^__label__([A-Za-z0-9_]+)")[0]
    df["text"]  = raw.str.replace(r"^__label__[A-Za-z0-9_]+\s*", "", regex=True).str.strip()

    # Binary: OFF=1, NOT=0
    df["label"] = df["Label"].str.upper().map({"NOT": 0, "OFF": 1})

    df = df[df["text"].str.len() > 0].copy()
    return df

tweet_train = parse_fasttext(ds_tw["train"].to_pandas())
tweet_val   = parse_fasttext(ds_tw["validation"].to_pandas())
tweet_test  = parse_fasttext(ds_tw["test"].to_pandas())

print("Train labels:\n", tweet_train["Label"].value_counts(dropna=False).to_string())
print("\nTrain binary:\n", tweet_train["label"].value_counts(dropna=False).to_string())
print("\nSample:\n", tweet_train[["Label","label","text"]].head(3).to_string(index=False))
```

```python
import pandas as pd
from pathlib import Path

# CSV yolu (senin yapına göre doğru)
news_path = Path("HateTargetBERT-TR") / "data" / "TurkishHatePrintCorpus.csv"

news_df = pd.read_csv(news_path)

# Temizlik
news_df["text"] = news_df["text"].astype(str).str.strip()
news_df = news_df[news_df["text"].str.len() > 0].copy()

# Label → binary
news_df["label"] = news_df["Label"].map({
    "nonhateful": 0,
    "hateful": 1
})

# Son hali
news_ready = news_df[["text", "label"]].copy()

print("News label distribution:")
print(news_ready["label"].value_counts().to_string())
print("\nSample rows:")
print(news_ready.head(3).to_string(index=False))
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
import evaluate
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer

metric_f1  = evaluate.load("f1")
metric_acc = evaluate.load("accuracy")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    return {
        "accuracy": metric_acc.compute(predictions=preds, references=labels)["accuracy"],
        "f1_macro": metric_f1.compute(predictions=preds, references=labels, average="macro")["f1"],
    }

train_df, val_df = train_test_split(
    news_ready,
    test_size=0.2,
    random_state=42,
    stratify=news_ready["label"]
)

model_name = "dbmdz/bert-base-turkish-cased"
tokenizer = AutoTokenizer.from_pretrained(model_name)

def tok(batch):
    return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=128)

train_ds = Dataset.from_pandas(train_df[["text","label"]].reset_index(drop=True)).map(tok, batched=True).remove_columns(["text"])
val_ds   = Dataset.from_pandas(val_df[["text","label"]].reset_index(drop=True)).map(tok, batched=True).remove_columns(["text"])
train_ds.set_format("torch"); val_ds.set_format("torch")

model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=2)

args = TrainingArguments(
    output_dir="run_news_only",
    eval_strategy="epoch",
    save_strategy="no",
    learning_rate=2e-5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=32,
    num_train_epochs=2,
    weight_decay=0.01,
    logging_steps=50,
    report_to="none",
)

trainer = Trainer(
    model=model,
    args=args,
    train_dataset=train_ds,
    eval_dataset=val_ds,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics,
)

trainer.train()
res = trainer.evaluate()
res
```
